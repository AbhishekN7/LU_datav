```python
from numpy.random import randn,randint,uniform,sample
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
x
```

```python
plt.plot(x,y,'go--', linewidth=2,markersize=8)
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Waveform')
plt.plot(x,y,color='b')
```

```python
plt.plot(x,y,linestyle='dashed',linewidth=2,markersize=10)
```
